```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Data Info:

The examined group comprised kernels belonging to three different varieties of wheat: Kama, Rosa and Canadian, 70 elements each, randomly selected for
the experiment. High quality visualization of the internal kernel structure was detected using a soft X-ray technique. It is non-destructive and considerably cheaper than other more sophisticated imaging techniques like scanning microscopy or laser technology. The images were recorded on 13x18 cm X-ray KODAK plates. Studies were conducted using combine harvested wheat grain originating from experimental fields, explored at the Institute of Agrophysics of the Polish Academy of Sciences in Lublin.

The data set can be used for the tasks of classification and cluster analysis.

Link:https://archive.ics.uci.edu/dataset/236/seeds

```python
cols=['area','perimeter','compactness','length','widht','asymmetry','groove','class']
dp=pd.read_csv('seeds_dataset.txt',names=cols,sep='\s+') #any space is recognized as data separator
```

```python
dp.head()
```

```python
for i in range(len(cols)-1):
  for j in range(i+1,len(cols)-1):
    x_label=cols[i]
    y_label=cols[j]
    sns.scatterplot(x=x_label, y=y_label, data=dp, hue='class') #hue wil separate 3 different classes into 3 diff hues
    plt.show()
```

#Clustering

```python
from sklearn.cluster import KMeans
```

```python
#x and y just refer to any two choosen features for this clustering
x='compactness'
y='asymmetry'
x1=dp[[x,y]].values
```

```python
kmeans=KMeans(n_clusters=3).fit(x1)
```

```python
clusters=kmeans.labels_
```

```python

clusters
```

```python
dp['class'].values
```

creating cluster dataframe.

```python
cluster_dp=pd.DataFrame(np.hstack((x1,clusters.reshape(-1,1))), columns=[x,y,'class'])
#in this dataframe x value is x1 and y value is  clusters which is reshaped in according to x1 and name of columns are x,y,'class'
```

```python
#K means classes
sns.scatterplot(x=x, y=y, data=cluster_dp, hue='class')
plt.plot()
```

```python
#Original classes
sns.scatterplot(x=x, y=y, data=dp, hue='class')
plt.plot()
```

#Higher Dimensions Clustering

```python
x2=dp[cols[:-1]].values
```

```python
kmeans=KMeans(n_clusters=3).fit(x2)
clusters=kmeans.labels_
```

```python
cluster_dp_high=pd.DataFrame(np.hstack((x2,clusters.reshape(-1,1))), columns=dp.columns)
```

```python
#K means classes for higher dimnesions
sns.scatterplot(x=x, y=y, data=cluster_dp_high, hue='class')
plt.plot()
```

```python
#Original classes for higher dimnesions
sns.scatterplot(x=x, y=y, data=dp, hue='class')
plt.plot()
```

higher dimensions clustering is more closer to result as compared to above single dimension clustering

#PCA


```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) #keeping the top two dimensions
transformed_x=pca.fit_transform(x2) #x2 from higher dimnesion clustering all the values accept 'class'
```

```python
x2.shape
```

```python
transformed_x.shape #only 2 dimnesions
```

```python
transformed_x[:5]
```

```python
plt.scatter(transformed_x[:,0],transformed_x[:,1])
```

```python
kmeans_pca_dp=pd.DataFrame(np.hstack((transformed_x,clusters.reshape(-1,1))),columns=['pca1','pca2','class'])
```

```python
truth_pca_dp=pd.DataFrame(np.hstack((transformed_x,dp['class'].values.reshape(-1,1))),columns=['pca1','pca2','class'])
```

```python
#K means PCA classes
sns.scatterplot(x='pca1', y='pca2', data=kmeans_pca_dp, hue='class')
plt.plot()
```

```python
#Original PCA classes
sns.scatterplot(x='pca1', y='pca2', data=truth_pca_dp, hue='class')
plt.plot()
```
